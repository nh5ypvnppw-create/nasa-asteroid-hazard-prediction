```python
import pandas as pd

pd.set_option('display.max_columns', None)
data = pd.read_csv("/content/nasa.csv")
data.head(7)
```

```python
data.columns
```

```python
data
```

```python
selected_features = [
    'Absolute Magnitude',
    'Est Dia in KM(min)',
    'Est Dia in KM(max)',
    'Relative Velocity km per sec',
    'Miss Dist.(Astronomical)',
    'Minimum Orbit Intersection',
    'Eccentricity',
    'Semi Major Axis',
    'Inclination',
    'Orbital Period',
    'Perihelion Distance',
    'Aphelion Dist',
    'Mean Motion'
]

target = "Hazardous"

X = data[selected_features]
y = data[target]

print("Selected Features: (X)")
print(X.head())

print("target (y)")
print(y.head())


```

```python
X.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15, 10))
for i,feature in enumerate(selected_features, 1):
  plt.subplot(4,4,i)
  sns.boxplot(y=X[feature])
  plt.title(feature)

plt.tight_layout()
plt.show()
```

```python
from numpy import power
from sklearn.preprocessing import PowerTransformer, StandardScaler
power_transformer = PowerTransformer(method = "yeo-johnson", standardize = True)
X_transformed = power_transformer.fit_transform(X)
X_transformed = pd.DataFrame(X_transformed, columns = selected_features)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_transformed,
    y,
    test_size=0.2,
    random_state=42)
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)

```

```python
y_pred
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Accuracy Score:")
print(accuracy_score(y_test, y_pred))

print("Confusion Matrix", confusion_matrix(y_test, y_pred))
```

```python
param_dist ={
    'max_depth' : randint(3, 10),
    'learning_rate' : uniform(0.01, 0.3),
    'n_estimators' : randint(100, 200),
    'subsample' : uniform(0.6, 0.4),
}

random_search = RandomizedSearchCV(
    estimator = xgb.XGBClassifier(objective='binary:logistic', random_state=42),
    param_distributions = param_dist,
    n_iter = 10,
    scoring = 'neg_log_loss',
    n_jobs = -1,
    cv = 5,
    random_state = 42
)

random_search.fit(X_train, y_train)

best_params_random = random_search.best_params_
print("Best Parameters:", best_params_random)

best_model_random = xgb.XGBClassifier(**best_params_random, objective='binary:logistic', random_state=42)
best_model_random.fit(X_train, y_train)

y_pred_random = best_model_random.predict(X_test)
y_pred_proba_random = best_model_random.predict_proba(X_test)[:, 1]

print("Classification Report (Randomized Search):")
print(classification_report(y_test, y_pred_random))

print("Log Loss (Randomized Search):", log_loss(y_test, y_pred_proba_random))
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, log_loss, classification_report, confusion_matrix
from bayes_opt import BayesianOptimization
import xgboost as xgb
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
from bayes_opt import BayesianOptimization
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, log_loss
import numpy as np
import xgboost as xgb


def xgb_evaluate(max_depth, learning_rate, n_estimators, subsample):
    model = xgb.XGBClassifier(
        objective='binary:logistic',
        max_depth=int(max_depth),
        learning_rate=learning_rate,
        n_estimators=int(n_estimators),
        subsample=subsample,
        random_state=42,
        n_jobs=-1,
        use_label_encoder=False,
        eval_metric='logloss'
    )


    scores = cross_val_score(model, X_train, y_train, scoring='neg_log_loss', cv=5)
    return np.mean(scores)


param_bounds = {
    'max_depth': (3, 10),
    'learning_rate': (0.01, 0.3),
    'n_estimators': (100, 200),
    'subsample': (0.6, 1.0)
}


optimizer = BayesianOptimization(
    f=xgb_evaluate,
    pbounds=param_bounds,
    random_state=42,
    verbose=2
)


optimizer.maximize(init_points=5, n_iter=10)


best_params = optimizer.max['params']
best_params['max_depth'] = int(best_params['max_depth'])
best_params['n_estimators'] = int(best_params['n_estimators'])

print("Best Parameters (Bayesian Optimization):", best_params)


best_model = xgb.XGBClassifier(
    **best_params,
    objective='binary:logistic',
    random_state=42,
    n_jobs=-1,
    use_label_encoder=False,
    eval_metric="logloss"
)

best_model.fit(X_train, y_train)


y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)[:, 1]

print("\nClassification Report (Bayesian Optimization):")
print(classification_report(y_test, y_pred))

print("Log Loss (Bayesian Optimization):", log_loss(y_test, y_pred_proba))




```
